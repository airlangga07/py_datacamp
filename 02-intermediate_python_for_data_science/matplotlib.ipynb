```python
# Working with Hans Rosling data about life expectancy for each country and GDP per capita
import pandas as pd

gap = pd.read_csv('datasets/gapminder.csv')

# get the life expectancy
life_exp = gap['life_exp'].tolist()

# get the GDP per capita
gdp_cap = gap['gdp_cap'].tolist()
```

```python
# with basic line plot, the life expectancy and gdp per capita does not pretty and have any relation
plt.plot(gdp_cap, life_exp)
plt.show()
```

## Scatter Plot

```python
# if you are trying to find the correlation between two variables or more, 
# scatter plot is the better choice
plt.scatter(gdp_cap, life_exp)
plt.title('Without logarithmic x scale')
plt.show()

# a correlation will become clear when we display the GDP per capita 
# on a logarithmic scale
plt.scatter(gdp_cap, life_exp)
plt.title('Without logarithmic x scale')
plt.xscale('log')
plt.show()
```

## Histogram

```python
# to see how life expectancy in different countries is distributed let's create a histogram
# if not specified, the bins will set to 10 as default
plt.hist(life_exp)
plt.show()

# with 5 bins
plt.hist(life_exp, bins=5)
plt.show()

# with 10 bins
plt.hist(life_exp, bins=20)
plt.show()
```

## Customization

```python
import numpy as np

# set the size for each dot by the number of population * 2
pop = gap['population'].tolist()
np_pop = np.array(pop)

# dividing with 1 million to fix the problem displaying the points being too big.
np_pop = np_pop * 2 / 1000000

# set the color for each of the points
r = 'red'
g = 'green'
b = 'blue'
y = 'yellow'
k = 'black'

colors = [r, g, b, b, y, k, g, r, r, g, b, y, g, b, y, g, b, b, r, b, 
          y, b, b, y, r, y, b, b, b, y, b, g, y, g, g, b, y, y, b, y, 
          b, b, b, g, g, b, b, g, b, g, y, b, b, y, y, r, g, g, r, r, 
          r, r, g, r, g, y, r, r, b, r, r, r, r, b, b, b, b, b, r, b, 
          b, b, y, r, g, b, b, r, b, r, g, k, y, b, b, g, r, r, y, y, 
          y, r, g, g, y, b, g, b, b, r, b, g, b, r, g, g, b, b, g, r, 
          b, b, g, g, r, r, b, r, b, y, b, g, b, g, y, y, y, r, r, r, b, b]

# set the plot
plt.scatter(
    gdp_cap,   # x plot
    life_exp,  # y plot
    s=np_pop,  # size of the points
    c=colors,  # set the colors for each of the points (in array)
    alpha=0.8  # change the opacity for each of the points
)

plt.xscale('log')

# set the x, y, and title lable
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Life Expectancy (yrs)')
plt.title('World Development 2007')

# Definition of tick_val and tick_lab
tick_val = [1000,10000,100000]
tick_lab = ['1k','10k','100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

plt.show()
```
